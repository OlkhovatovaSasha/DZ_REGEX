## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
import re
text = ['_065мк 78', 'а258вы 99', '454259 65', '!№к546 ке', 'у874ав 66','ш547гн 96']

regex_text = r'([а-я]{1})(\d{3})[а-я]{2}\s(\d{2})'


for line in text:
    if re.findall(regex_text, line):
        namber = line[:6]
        region = line[7:]
        
        print('Номер: ' + namber + ' Регион: '+ region)
```

Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:   
+7(xxx)-xxx-xx-xx

Примеры правильных и неправильных форматов:
- +7 955 555-55-55;
- 955555555;
- 8(955)555-55-55;
- +7 955 555 55 55;
- 7(955) 555-55-55;

```python
import re
number_array = ['10 955 555-55-55', 'g555555555', '5(955)555-55-55', '+7 955 555 55 55', '8(955) 555-55-55','+7 955+555+55+55']
reg = r'(^(7|8|9))' 

for number_one in number_array: 
    number_clear = re.sub(r'(\+|\-|\s|[\(\)])','', number_one)
        
    if re.findall(reg, number_clear):
        
        if len(number_clear) == 11:
            namber_correct = '+7(' + number_clear[1:4] + ')-' + number_clear[4:7] + '-' + number_clear[7:9] + '-' + number_clear[9:]
            print(namber_correct)
        
        if len(number_clear) == 10:
            namber_correct = '+7(' + number_clear[0:3] + ')-' + number_clear[3:6] + '-' + number_clear[6:8] + '-' + number_clear[8:]
            print(namber_correct)
```
